```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import seaborn as sns
```

```python
data = [
    [590,   61,     9,      1,      0],
    [30,    90,     10,     0,      0],
    [12,    70,     400,    42,     0],
    [0,     31,     54,     200,    90],
    [0,     0,      12,     160,    700]
]
data = pd.DataFrame(data)
data
```

```python
class_names = [f'class {x}' for x in range(0, 5)]
class_names
```

```python
# No normalization
ax: plt.Axes = sns.heatmap(data, annot=True, fmt='.0f', cmap='viridis', xticklabels=class_names, yticklabels=class_names)
ax.set_title('Normalization: None')
```

```python
# True normalization
ax: plt.Axes = sns.heatmap(data.div(data.sum(axis=1), axis=0), annot=True, fmt='.2f', cmap='viridis', xticklabels=class_names, yticklabels=class_names)
ax.set_title('Normalization: True Labels')
```

```python
# Predicted normalization
ax: plt.Axes = sns.heatmap(data.div(data.sum(axis=0), axis=1), annot=True, fmt='.2f', cmap='viridis', xticklabels=class_names, yticklabels=class_names)
ax.set_title('Normalization: Predicted Labels')
```
