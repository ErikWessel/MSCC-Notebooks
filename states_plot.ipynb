```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.colors import LinearSegmentedColormap
```

```python
target_states = ['California', 'Florida', 'Montana', 'Washington', 'Texas']
```

```python
usa_states: gpd.GeoDataFrame = gpd.read_file('input/us_states_5m.geojson')
usa_states['color'] = usa_states['NAME'].isin(target_states).apply(lambda x: 1 if x else 0)
```

```python
cmap = LinearSegmentedColormap.from_list('states_cmap', [(0, 'lightgray'), (1, 'cornflowerblue')])
fig, ax = plt.subplots(figsize=(16, 16))
usa_states.plot(ax=ax, linewidth=0.5, edgecolor='white', cmap=cmap, column='color')
ax.set_ylim(24, 52)
ax.set_xlim(-128, -65)
ax.axis(False)
```
